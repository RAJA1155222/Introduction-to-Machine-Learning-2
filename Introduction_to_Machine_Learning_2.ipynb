Q1. Define overfitting and underfitting.

What are their consequences and how can they be mitigated?

🔹 Overfitting

Overfitting happens when a model learns the training data too well, including noise and random fluctuations.

Symptoms

Very high training accuracy

Poor test / validation accuracy

Consequences

Poor generalization

Unreliable predictions on new data

How to mitigate

Use more data

Apply regularization (L1 / L2)

Use dropout (for neural networks)

Reduce model complexity

Use early stopping

Use cross-validation

🔹 Underfitting

Underfitting happens when a model is too simple to capture the underlying pattern in the data.

Symptoms

Poor training accuracy

Poor test accuracy

Consequences

The model fails to learn meaningful relationships

High prediction error everywhere

How to mitigate

Use a more complex model

Add more relevant features

Train for more epochs

Reduce regularization strength

✅ Q2. How can we reduce overfitting? (Brief)

Common techniques to reduce overfitting:

Use more training data

Use dropout (deep learning)

Reduce model complexity

Use early stopping

✅ Q4. Explain the bias–variance tradeoff.
🔹 Bias

Error caused by wrong assumptions in the learning algorithm.

High bias → model is too simple.

🔹 Variance

Error caused by high sensitivity to training data.

High variance → model is too complex.

🔹 Relationship

There is a tradeoff:

Increasing model complexity → lowers bias but increases variance

Decreasing model complexity → increases bias but lowers variance

🔹 Effect on performance
Case	Bias	Variance	Result
Underfitting	High	Low	Poor training & test performance
Overfitting	Low	High	Good training, poor test performance
Good model	Balanced	Balanced	Good generalization
✅ Q5. How to detect overfitting and underfitting?

The most common way is by comparing:

training performance

validation / test performance

🔹 Overfitting detection

Training accuracy is very high

Validation accuracy is much lower

or

Training loss is very low

Validation loss is high

🔹 Underfitting detection

Training accuracy is low

Validation accuracy is also low

or

Both training and validation losses are high

Learning curve method

Plot:

training loss vs epochs

validation loss vs epochs

This clearly shows whether the model is overfitting or underfitting.

✅ Q7. What is regularization and how does it prevent overfitting?

Regularization is a technique that adds a penalty to the loss function in order to limit model complexity.

Its goal is to prevent the model from learning overly complex patterns.

🔹 How regularization helps

discourages large weights

makes the model smoother

reduces variance

improves generalization

🔹 Common regularization techniques
1. L1 regularization (Lasso)

Adds absolute value of weights to the loss.

Effect:

encourages sparsity

some weights become zero

performs feature selection

2. L2 regularization (Ridge / weight decay)

Adds squared weights to the loss.

Effect:

shrinks weights

stabilizes learning

reduces sensitivity to noise

3. Dropout (deep learning)

Randomly disables neurons during training.

Effect:

prevents co-adaptation

acts like training many smaller networks

4. Early stopping

Stops training when validation performance stops improving.

Effect:

prevents learning noise

reduces effective model complexity

5. Data augmentation

Artificially increases training data.

Effect:

reduces overfitting

improves robustness
